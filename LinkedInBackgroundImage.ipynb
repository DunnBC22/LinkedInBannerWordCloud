```python
#Source: Duong Vu (https://www.datacamp.com/community/tutorials/wordcloud-python)

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import text
words=[]

with open('/Users/briandunn/wordcloud/WordsToInclude.txt', 'r') as file:
    words=[file.read()]

print(words)
```

```python
wordcloud = WordCloud(background_color="beige", width=1584, height=396).generate(' '.join(words))
# Generate plot
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
#plt.draw()
```

```python
#output the image
wordcloud.to_file("/Users/briandunn/Documents/Output.png")
```
